```python
import pandas as pd
```

```python
data = [1,2,3,4,5]
s=pd.Series(data)
print(s)
```

```python
# Create a searies from dictionary
data={'a':1, 'b':2, 'c':3}
s_dict=pd.Series(data)
print(s_dict)
```

```python
data = [10, 20, 30]
index = ['a','b', 'c']

pd.Series(data, index=index)
```

```python
# Dataframe
# create a dataframe from a dictionary of list

data={
    'Name':['Krish','John','Jack'],
    'Age':[25, 30, 45],
    'City':['Bangalore','New York','Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```
